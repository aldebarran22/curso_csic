## Gráficas con la librería matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Cambiar el tamaño del área de trazado
mpl.rcParams['figure.figsize']=(12,7)
```

```python
x = np.arange(1,10,0.25)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)
```

```python
plt.subplot2grid((2,2),(0,0))
plt.plot(x, y, color="blue")
plt.title('sin(x)')
plt.grid(True)

plt.subplot2grid((2,2),(0,1))
plt.plot(x, y2, color="green")
plt.title('cos(x)')

plt.subplot2grid((2,2),(1,0), colspan=2)
plt.plot(x, y3, color="red")
plt.title('Exp(x)')
plt.suptitle('Funciones')
plt.show()
```

```python
plt.plot(x, y, color="blue", label='sin(x)')
plt.plot(x, y2, color="red", label='cos(x)')
#plt.plot(x, y3, color="orange", label='exp(x)')
plt.legend()
plt.grid(True,ls=':',color='r')
plt.title('Sin / Cos')
plt.text(2,1,"Max sin(x)")
plt.text(5,-1,"Min sin(x)")
plt.annotate("sin(x) = cos(x)", xytext=(6, 0.75), xy=(7, 0.7), \
             ha='right',\
             va='bottom',arrowprops={'color':'black','shrink':0.02})
plt.show()
```
